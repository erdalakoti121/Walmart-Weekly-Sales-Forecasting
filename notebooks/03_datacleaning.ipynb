```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.float_format", "{:,.2f}".format)
```

```python
train = pd.read_csv("../data/raw/train.csv")
stores = pd.read_csv("../data/raw/stores.csv")
features = pd.read_csv("../data/raw/features.csv")
```

```python
merged_df = train.merge(stores, on="Store", how="left")

final_merge = merged_df.merge(
    features,
    on=["Store", "Date"],
    how="left"
)
final_merge.info()
```

```python
final_merge.info()
```

```python
final_merge.isna().sum().sort_values(ascending=False)
```

```python
final_merge[
    ["Date", "MarkDown1", "MarkDown2", "MarkDown3", "MarkDown4", "MarkDown5"]
].head(20)
```

```python
final_merge["MarkDown1"] = final_merge["MarkDown1"].fillna(0)
final_merge["MarkDown2"] = final_merge["MarkDown2"].fillna(0)
final_merge["MarkDown3"] = final_merge["MarkDown3"].fillna(0)
final_merge["MarkDown4"] = final_merge["MarkDown4"].fillna(0)
final_merge["MarkDown5"] = final_merge["MarkDown5"].fillna(0)
```

```python
final_merge.isna().sum().sort_values(ascending=False)
```

```python
final_merge.duplicated().sum()
```

```python
final_merge["Date"] = pd.to_datetime(final_merge["Date"])
final_merge.info()
```

```python
final_merge.to_csv(
    "../data/cleaned/final_cleaned.csv",
    index=False
)
```
