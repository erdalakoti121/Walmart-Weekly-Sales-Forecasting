```python
import pandas as pd
import numpy as np
```

```python
final_data = pd.read_csv("../data/cleaned/final_cleaned.csv")
```

```python
final_data["Type"].unique()
```

```python
final_data["Type"] = final_data["Type"].str.strip()
```

#feature enngineering
type to numeric vlaues 

```python
final_data["Type"] = final_data["Type"].replace({
    "A": 0,
    "B": 1,
    "C": 2
})
```

```python
final_data["Type"].unique()
```

feature engineering 

```python
final_data["IsHoliday_x"] = final_data["IsHoliday_x"].astype(int)
```

```python
final_data["IsHoliday_x"].unique()
```

```python
final_data["Date"] = pd.to_datetime(final_data["Date"])
```

```python
final_data["year"]= final_data["Date"].dt.year
final_data["month"]= final_data["Date"].dt.month
final_data["Quarter"]= final_data["Date"].dt.quarter
final_data["week"]= final_data["Date"].dt.isocalendar().week
final_data["DayOfWeek"] = final_data["Date"].dt.dayofweek
```
